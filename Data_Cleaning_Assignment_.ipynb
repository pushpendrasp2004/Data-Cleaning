- Data cleaning (also called data cleansing) is the process of identifying, correcting, and removing errors, inconsistencies, inaccuracies, and irrelevant information from raw data to improve its quality and reliability before analysis.
- importance: Ensures accuracy and correctness of analytical results

Improves consistency and completeness of data

Reduces bias and misleading conclusions

Enhances model performance in statistical and machine learning tasks

Saves time and cost by avoiding rework and incorrect decisions

- Consequences of Analyzing Unclean or Messy Data

Analyzing unclean data can lead to serious technical and business problems:

Incorrect Results
Errors, duplicates, or missing values can distort analysis outcomes.

Misleading Insights
Inconsistent or biased data may lead to wrong patterns or trends.

Poor Decision-Making
Decisions based on faulty data can cause financial losses and operational failures.

Low Model Accuracy
Machine learning models trained on noisy data produce unreliable predictions.

Loss of Credibility
Stakeholders may lose trust in reports and analytical systems.

Compliance and Legal Risks
Inaccurate data may violate data governance or regulatory standards.

-Common Steps Involved in Cleaning and Organizing Data
Step 1: Data Inspection and Understanding

Review dataset structure, data types, and summary statistics

Identify missing values, outliers, and anomalies

Step 2: Handling Missing Data

Remove rows or columns with excessive missing values

Replace missing values using:

Mean / median / mode

Forward or backward filling

Predictive imputation

Step 3: Removing Duplicate Records

Identify and eliminate duplicate entries to avoid data redundancy.

Step 4: Correcting Inconsistencies

Standardize formats (e.g., date, text case, units)

Fix spelling errors and inconsistent category labels

Step 5: Handling Outliers

Detect extreme values using statistical methods

Remove or cap outliers if they are errors or irrelevant

Step 6: Data Validation

Ensure values fall within acceptable ranges

Verify logical relationships between variables

Step 7: Data Transformation and Organization

Normalize or scale numerical data

Encode categorical variables

Rename columns for clarity

Sort and structure data for analysis

In the Settings tab:

Minimum → 1

Maximum → 10

Click on the Input Message tab

Check “Show input message when cell is selected”

Enter:

Title: Quantity Input

Input Message:

Please enter a whole number between 1 and 10.


This message appears when the user selects a Quantity cell.

Step 6: Set Error Alert Message

Click on the Error Alert tab

Check “Show error alert after invalid data is entered”

Set:

Style: Stop

Title: Invalid Input

Error Message:

Invalid input! The quantity must be a whole number between 1 and 10.


This message appears if the user enters a number less than 1, greater than 10, or a decimal value.

Step 7: Apply the Validation

Click OK

Data Validation is now active on the Quantity

 - a. Difference Between TODAY() and NOW()
Function	What it Returns	Example Use
TODAY()	Current date only	Tracking deadlines, due dates
NOW()	Current date and time	Logging time-stamped entries
Examples:

=TODAY() → 06/09/2026

=NOW() → 06/09/2026 11:45 AM

Usage:

Use TODAY() when only the date matters

Use NOW() when both date and time are required

If cell A1 = 2025-06-10

=A1 - TODAY()


Result:
